### MeanShiftClustering Model

https://www.kaggle.com/datasets/vinicius150987/titanic3

finds its own number of clusters

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')

from sklearn.cluster import MeanShift

from sklearn import preprocessing
from sklearn.model_selection import train_test_split
```

```python
# Read the data
df = pd.read_excel('Data/titanic.xls')

# Make a copy of the df
original_df = pd.DataFrame.copy(df)

df.head()
```

```python
df.drop(['body','name'], 1, inplace=True)
df.fillna(0, inplace=True)

# Convert non-numerical data into numerical data
def handle_non_numerical_data(df):
    columns = df.columns.values
    for column in columns:
        text_digit_vals = {}
        def convert_to_int(val):
            return text_digit_vals[val]

        if df[column].dtype != np.int64 and df[column].dtype != np.float64:
            column_contents = df[column].values.tolist()
            unique_elements = set(column_contents)
            x = 0
            for unique in unique_elements:
                if unique not in text_digit_vals:
                    text_digit_vals[unique] = x
                    x+=1

            df[column] = list(map(convert_to_int, df[column]))

    return df
```

```python
df = handle_non_numerical_data(df)
df.head()
```

```python
# Separate into features and labels
X = np.array(df.drop(['survived'], 1).astype(float))

# Scale the features: the other results were not good
X = preprocessing.scale(X)

y = np.array(df['survived'])
```

```python
# Clustering algorithm
clf = MeanShift()
clf.fit(X)
```

```python
cluster_centers = clf.cluster_centers_
labels = clf.labels_
```

```python
# Add the labels to the original df
original_df['cluster_grous'] = labels
```
