### LogReg model

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import random
```

```python
X1 = []
X2 = []
Class = []

for i in range(1000):
    x1 = np.random.uniform(-2,2)
    x2 = np.random.uniform(-2,2)
    
    X1.append(x1)
    X2.append(x2)
    
    if x1**2 + x2**2 < 1:
        Class.append(0)
    else:
        Class.append(1)
    
```

```python
df = pd.DataFrame()
df['X1'] = X1
df['X2'] = X2
df['Class'] = Class
```

```python
for i in range(1000):
    if df.iloc[i]['Class'] == 0:
        plt.scatter(df.iloc[i]['X1'], df.iloc[i]['X2'], s=2, c='red')
    else:
        plt.scatter(df.iloc[i]['X1'], df.iloc[i]['X2'], s=2, c='blue')
        
```

```python
# Separate Xthe data
X = np.array(df.drop(['Class'], 1))
y = np.array(df['Class'])

# Separate X values into train-test and then prediction 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1)
```

```python
# Define the classifier
clf = LogisticRegression(random_state=0)
clf.fit(X_train, y_train)
```

```python
# Test the classifier
accuracy = clf.score(X_test, y_test)
accuracy
```
