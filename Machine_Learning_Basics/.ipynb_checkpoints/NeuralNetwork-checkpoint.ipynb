### NN model

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
#from matplotlib import style
#style.use('ggplot')
```

```python
X1 = []
X2 = []
Class = []

for i in range(1000):
    x1 = np.random.uniform(-2,2)
    x2 = np.random.uniform(-2,2)
    
    X1.append(x1)
    X2.append(x2)
    
    if x1**2 + x2**2 < 1:
        Class.append(0)
    else:
        Class.append(1)
    
```

```python
df = pd.DataFrame()
df['X1'] = X1
df['X2'] = X2
df['Class'] = Class
```

```python
for i in range(1000):
    if df.iloc[i]['Class'] == 0:
        plt.scatter(df.iloc[i]['X1'], df.iloc[i]['X2'], s=2, c='red')
    else:
        plt.scatter(df.iloc[i]['X1'], df.iloc[i]['X2'], s=2, c='blue')
        
```
