### This notebook creates and analyses randomly generated time series data

Source: https://www.timescale.com/blog/how-to-work-with-time-series-in-python/

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Generate random time series data
LENGTH = 100

np.random.seed(42)
dates = pd.date_range(start='2022-01-01', periods=LENGTH, freq='D')
values = np.random.randn(LENGTH).cumsum()

# Create a dataframe from the generated data
data = pd.DataFrame({'date': dates, 'value': values})

# Set the 'date' column as the index
data.set_index('date', inplace=True)
```

```python
# Plot the time-series data
plt.plot(data.index, data['value'])
plt.xlabel('Time')
plt.ylabel('Value')
plt.xticks(rotation = 45)
plt.title('Time Series Data')
plt.show()
```

```python
# Plot the time-series data
plt.plot(dates, values)
plt.xlabel('Date')
plt.ylabel('Value')
plt.xticks(rotation = 45)
plt.title('Time Series Data')
plt.show()
```
