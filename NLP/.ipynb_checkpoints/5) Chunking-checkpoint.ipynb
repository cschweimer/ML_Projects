### Chunking

text is split up by part of speech tags

split text by nouns to find out meaning $\rightarrow$ noun phrases

```python
import nltk
from nltk.corpus import state_union
from nltk.tokenize import PunktSentenceTokenizer
```

```python
train_text = state_union.raw('2005-GWBush.txt')
sample_text = state_union.raw('2006-GWBush.txt')
```

```python
custom_sent_tokenizer = PunktSentenceTokenizer(train_text)

tokenized = custom_sent_tokenizer.tokenize(sample_text)
```

```python
def process_content():
    try:
        for i in tokenized:
            words = nltk.word_tokenize(i)
            tagged = nltk.pos_tag(words)
            
            # Look for adverb -> RB and noun -> NNP
            chunkGram = r'''Chunk: {<RB.?>*<VB.?>*<NNP>+<NN>?} '''
            
            chunkParser = nltk.RegexpParser(chunkGram)
            chunked = chunkParser.parse(tagged)
            
            print(chunked)
            # Plot the result for better visual
            chunked.draw()
            
    except Exception as e:
        print(str(e))
        
```

```python
process_content()
```

### Chinking

```python
def process_content():
    try:
        for i in tokenized:
            words = nltk.word_tokenize(i)
            tagged = nltk.pos_tag(words)
            
            # "anything zero or more repititions"
            chunkGram = r'''Chunk: {<.*>+} 
            # Keep out any verb, and other
                                   }<VB.?|IN|DT|TO>+{'''
            
            chunkParser = nltk.RegexpParser(chunkGram)
            chunked = chunkParser.parse(tagged)
            
            print(chunked)
            # Plot the result for better visual
            chunked.draw()
            
    except Exception as e:
        print(str(e))
        
```

```python
process_content()
```
